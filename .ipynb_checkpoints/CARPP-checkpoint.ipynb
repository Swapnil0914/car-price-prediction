```python
df['year']=2020-df['year']
```

```python
df=pd.get_dummies(df,columns=['transmission','fuelType'],drop_first=True)
```

```python
df.head()
```

```python
df=df.drop(['Make','model'],axis=1)
```

```python
df.head()
```

```python
X=df.drop('price',axis=1)
Y=df['price']
print("X size ",X.shape)
print("Y size ",Y.shape)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.40, random_state=42)  
```

```python
rf = ensemble.RandomForestRegressor(n_estimators=200,
                               
                               max_features=6,
                               n_jobs=-1,
                               random_state=1)
rf.fit(X_train,Y_train)
Y_predict=rf.predict(X_test)
metrics.r2_score(Y_test,Y_predict)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(Y_predict,Y_test ))  
print('Mean Squared Error:', metrics.mean_squared_error(Y_predict,Y_test))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Y_test, Y_predict)))
```

```python
dtfr = pd.DataFrame({'Actual': Y_test, 'Predicted': Y_predict})
dtfr
```

```python
dtfr1 = dtfr.head(50)
dtfr1.plot(kind='bar',figsize=(16,10))
pt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
pt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
pt.show()
```

```python
pt.figure(figsize=(15,10))
pt.scatter(Y_test,Y_predict)
pt.xlabel('Actual')
pt.ylabel('Predicted')
pt.title('ACTUAL vs PREDICTED')
```
